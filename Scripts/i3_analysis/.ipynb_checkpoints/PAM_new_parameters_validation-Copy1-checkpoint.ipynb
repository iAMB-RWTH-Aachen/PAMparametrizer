```python
import os 
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

if os.path.split(os.getcwd())[1] == 'Visualization':
    os.chdir('../..')
    
from Scripts.pam_generation import setup_ecoli_pam
from Scripts.Visualization.PAMparametrizer_progress_cleaned_figure import run_simulations
```

```python
RXN_NAME_MAPPER = {'EX_ac_e': 'Acetate secretion [$mmol_{ac}/g_{CDW}/h$]',
                   'EX_glc__D_e': 'Glucose uptake [$mmol_{glc}/g_{CDW}/h$]',
                   'EX_co2_e': '$CO_2$ secretion [$mmol_{CO_2}/g_{CDW}/h$]',
                   'EX_o2_e': 'Oxygen uptake [$mmol_[{O_2}/g_{CDW}/h$]',
                   'BIOMASS_Ec_iML1515_core_75p37M': 'Growth rate [$h^{-1}$]'}
MARKERS = ['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
ref_data_path = os.path.join('Data', 'Ecoli_phenotypes', 'Ecoli_phenotypes_py_rev.xls')
```

## Setup the model and get the reference data

```python
ecoli_pam = setup_ecoli_pam(pam_data_file_path = os.path.join(
    'Data', 'proteinAllocationModel_iML1515_EnzymaticData_py.xls'))
ref_data_df = pd.read_excel(ref_data_path, sheet_name='Yields')
```

calculate yields and RQ for all the datapoints

```python
ref_data_df['RQ'] = ref_data_df.EX_co2_e/ref_data_df.EX_o2_e.abs()
ref_data_df['Yxs'] = ref_data_df.BIOMASS_Ec_iML1515_core_75p37M.abs()/ref_data_df.EX_glc__D_e.abs()#biomass yield
ref_data_df['Yas'] = ref_data_df.EX_ac_e.abs()/ref_data_df.EX_glc__D_e.abs() #acetate yield
```

## Perform simulations

```python
def run_simulations(pamodel, substrate_rates, rxn_to_validate = RXNS_TO_VALIDATE):
    result_df = pd.DataFrame(columns= rxn_to_validate)

    for substrate in substrate_rates:
        pamodel.change_reaction_bounds(rxn_id=config.GLUCOSE_EXCHANGE_RXNID,
                                       lower_bound=substrate, upper_bound=0)
        print('Running simulations with ', substrate, 'mmol/g_cdw/h of substrate going into the system')
        pamodel.optimize()
        if pamodel.solver.status == 'optimal' and pamodel.objective.value>0:
            results_row = []
            for rxn in rxn_to_validate:
                results_row += [pamodel.reactions.get_by_id(rxn).flux]

            result_df.loc[len(result_df)] = [substrate] + results_row
    return result_df
```

```python
substrate_uptake_rates = np.arange(-11, 1, 1)

fluxes = run_simulations(ecoli_pam, substrate_uptake_rates)
```

## Plot the simulations

```python
def plot_simulations(valid_data, fluxes, fontsize:int = 12):
    print(RXN_NAME_MAPPER)    
    reactions_to_plot = list(RXN_NAME_MAPPER.keys())
    
    # plot flux changes with glucose uptake
    fig, ax = plt.subplots(1, dpi=100)
    
    cmap = mpl.colormaps['viridis']
    # Take colors at regular intervals spanning the colormap.
    colors = cmap(np.linspace(0, 1, len(reactions_to_plot)))
    
    for i,r in enumerate(reactions_to_plot):
        x = [abs(glc) for glc in valid_data['EX_glc__D_e']]
        y = [abs(data) for data in valid_data[r]]
        
        if 'BIOMASS' in r:
            ax2 = ax.twinx()
            ax2.set_ylabel('Growth rate [$h^(-1)$]', fontsize= fontsize)
            ax2.scatter(x, y,
                       color=colors[i], marker=MARKERS[i], s=20,
                       facecolors=None, zorder=0,
                       label=RXN_NAME_MAPPER[r])
            ax2.plot([abs(f['EX_glc__D_e']) for f in fluxes], [abs(f[r]) for f in fluxes], linewidth=3,
                       zorder=5, color=colors[i], label = RXN_NAME_MAPPER[r])
            continue
        # plot data
        ax.set_ylabel('Flux rate [$mmol/g_{CDW}/h$]', fontsize= fontsize)
        ax.scatter(x, y,
                       color=colors[i], marker=MARKERS[i], s=20,
                       facecolors=None, zorder=0,
                       label=RXN_NAME_MAPPER[r])
        ax.plot([abs(f['EX_glc__D_e']) for f in fluxes], [abs(f[r]) for f in fluxes], linewidth=3,
                       zorder=5, color=colors[i], label = RXN_NAME_MAPPER[r])
        ax.set_xlabel(RXN_NAME_MAPPER['EX_glc__D_e'], fontsize= fontsize)
        ax.tick_params(axis='both',labelsize = fontsize)
#         ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
    plt.legend()
    fig.tight_layout()
    fig.show()

    return fig, ax
```

```python
plot_simulations(ref_data_df, fluxes)
```

```python
import os 
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec


if os.path.split(os.getcwd())[1] == 'i1_preprocessing':
    os.chdir('../..')
    
from Scripts.pam_generation import setup_ecoli_pam
from Scripts.pam_generation_uniprot_id import set_up_ecoli_pam as setup_ecoli_pam_up

from Modules.utils.sector_config_functions import run_simulations
```

```python
RXN_NAME_MAPPER = {'EX_ac_e': 'Acetate secretion',
                   'EX_glc__D_e': 'Glucose uptake',
                   'EX_co2_e': '$CO_2$ secretion',
                   'EX_o2_e': 'Oxygen uptake',
                   'BIOMASS_Ec_iML1515_core_75p37M': 'Growth rate [$h^{-1}$]'}
MARKERS = ['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
RESULT_FIG_FILE = os.path.join('Figures', 'Figure1_PAM_performance.png')
ref_data_path = os.path.join('Data', 'Ecoli_phenotypes', 'Ecoli_phenotypes_py_rev.xls')
```

```python
def calculate_yields_rq(df):
    df['RQ'] = df.EX_co2_e/df.EX_o2_e.abs()
    df['Yxs'] = df.BIOMASS_Ec_iML1515_core_75p37M.abs()/df.EX_glc__D_e.abs()#biomass yield
    df['Yas'] = df.EX_ac_e.abs()/df.EX_glc__D_e.abs() #acetate yield
    return df
```

## Setup the model and get the reference data

```python
ecoli_pam = setup_ecoli_pam(pam_data_file_path = os.path.join(
    'Data', 'proteinAllocationModel_iML1515_EnzymaticData_py.xls'), sensitivity = False)
ecoli_pam_new = setup_ecoli_pam(pam_data_file_path = os.path.join(
    'Data', 'proteinAllocationModel_iML1515_EnzymaticData_new.xls'), sensitivity = False)
ecoli_pam_uniprot = setup_ecoli_pam_up(pam_info_file = os.path.join(
    'Results', '1_preprocessing', 'proteinAllocationModel_iML1515_EnzymaticData_240730.xlsx'), sensitivity = False)
ref_data_df = pd.read_excel(ref_data_path, sheet_name='Yields')
```

calculate yields and RQ for all the datapoints

```python
ref_data_df = calculate_yields_rq(ref_data_df)
```

## Perform simulations

```python
substrate_uptake_rates = np.arange(-11, 1, 1)

print('Running simulations for the PAM with the old parameters')
fluxes = run_simulations(ecoli_pam, substrate_uptake_rates)
# fluxes = calculate_yields_rq(fluxes)
print('\nRunning simulations for the PAM with the new parameters')
fluxes_new = run_simulations(ecoli_pam_new, substrate_uptake_rates)
# fluxes_new = calculate_yields_rq(fluxes_new)

print('\nRunning simulations for the PAM with the new parameters and complex protein-reaction associations')
fluxes_uniprot = run_simulations(ecoli_pam_uniprot, substrate_uptake_rates)
```

## Plot the simulations

```python
def plot_simulations(valid_data, fluxes, fig=None, grdspc=None, fontsize: int = 12, legend=True, title=None):
    reactions_to_plot = [rxn for rxn in RXN_NAME_MAPPER.keys() if rxn != 'EX_glc__D_e']

    # plot flux changes with glucose uptake
    if not fig is None:
        gs = gridspec.GridSpecFromSubplotSpec(2, 1, hspace=0, subplot_spec=grdspc)
        ax = fig.add_subplot(gs[0, 0])
        ax_yield = fig.add_subplot(gs[1, 0], sharex=ax)
    else:
        fig, axs = plt.subplots(2, dpi=100)
        fig.set_figwidth(15)
        fig.set_figheight(15)
        ax = axs.flatten()[0]
        ax_yield = axs.flatten()[1]

    # Use colormap to generate colors
    cmap = plt.cm.get_cmap('viridis', len(reactions_to_plot) + 3)  # +3 for additional plots
    colors = cmap(np.linspace(0, 1, len(reactions_to_plot) + 3))

    for i, r in enumerate(reactions_to_plot):
        x = [abs(glc) for glc in valid_data['EX_glc__D_e']]
        y = [abs(data) for data in valid_data[r]]
        xsim = [abs(f['EX_glc__D_e']) for f in fluxes]
        ysim = [abs(f[r]) for f in fluxes]

        if 'BIOMASS' in r:
            ax2 = ax.twinx()
            ax2.set_ylabel('Growth rate [$h^(-1)$]', fontsize=fontsize)
            s = ax2.scatter(x, y, color=colors[i], marker=MARKERS[i], s=20, zorder=0)
            l = ax2.plot(xsim, ysim, linewidth=3, zorder=5, color=colors[i])
            ax.plot([], [], marker=MARKERS[i], color=colors[i], label=RXN_NAME_MAPPER[r])  # only legend handle
            continue

        # plot data
        ax.set_ylabel('Flux rate [$mmol/g_{CDW}/h$]', fontsize=fontsize)
        s = ax.scatter(x, y, color=colors[i], marker=MARKERS[i], s=20, zorder=0)
        l = ax.plot(xsim, ysim, linewidth=3, zorder=5, color=colors[i])
        ax.plot([], [], marker=MARKERS[i], color=colors[i], label=RXN_NAME_MAPPER[r])  # only legend handle
        ax.set_xlabel(RXN_NAME_MAPPER['EX_glc__D_e'], fontsize=fontsize)
        ax.tick_params(axis='both', labelsize=fontsize)

    # Add respiratory coefficients, acetate yield, and biomass yield on ax_yield
    rqsim = [abs(f['EX_co2_e'] / f['EX_o2_e']) for f in fluxes]
    yassim = [abs(f['EX_ac_e'] / f['EX_glc__D_e']) for f in fluxes]
    yxssim = [abs(f['BIOMASS_Ec_iML1515_core_75p37M'] / f['EX_glc__D_e']) for f in fluxes]

    ax_yield.plot(xsim, rqsim, color=colors[i + 1], linewidth=3, zorder=0, label='Respiratory coefficient (RQ)')
    ax_yield.scatter(x, [abs(data) for data in valid_data['RQ']], color=colors[i + 1], marker=MARKERS[i], s=20, zorder=0)
    ax_yield.plot(xsim, yassim, color=colors[i + 2], linewidth=3, zorder=0, label='Acetate yield on substrate [$Y_{a/s}$]')
    ax_yield.scatter(x, [abs(data) for data in valid_data['Yas']], color=colors[i + 2], marker=MARKERS[i - 1], s=20, zorder=0)
    ax_yield.plot(xsim, yxssim, color=colors[i + 3], linewidth=3, zorder=0, label='Biomass yield on substrate [$Y_{x/s}$]')
    ax_yield.scatter(x, [abs(data) for data in valid_data['Yxs']], color=colors[i + 3], marker=MARKERS[i - 2], s=20, zorder=0)

    ax_yield.set_xlabel(RXN_NAME_MAPPER['EX_glc__D_e'], fontsize=fontsize)
    ax_yield.set_ylabel('Yield/RQ', fontsize=fontsize)
    ax_yield.set_ylim([-0.1, 1.3])
    ax_yield.tick_params(axis='both', labelsize=fontsize)

    # Collect legend handles and labels
    handles_ax, labels_ax = ax.get_legend_handles_labels()
    handles_yield, labels_yield = ax_yield.get_legend_handles_labels()
    combined_handles = handles_ax + handles_yield
    combined_labels = labels_ax + labels_yield

    # Create a combined legend
    if legend:
        ax.legend(combined_handles, combined_labels, loc='upper left')
        ax_yield.legend()

    if isinstance(title, str):
        ax.set_title(title, fontsize=fontsize * 1.5)

    fig.tight_layout()
    fig.show()

    return fig, combined_handles, combined_labels
```

```python
plot_simulations(ref_data_df, fluxes)
```

```python
plot_simulations(ref_data_df, fluxes_new)
```

```python
plot_simulations(ref_data_df, fluxes_uniprot)
```

## Create one big figure to save

```python
fig = plt.figure()
fontsize = 15
annotation = ['A', 'B', 'C']
titles = ['Alter et al. (2021)', 'novel sectors', 'novel sectors + GPRAs']

fig.set_figwidth(20)
fig.set_figheight(8)

gs0 = gridspec.GridSpec(1, 3, figure=fig)
for i, flux_df in enumerate([fluxes, fluxes_new, fluxes_uniprot]):
    gs = gs0[0,i]
    fig, handles, labels = plot_simulations(ref_data_df, flux_df, fig, gs, title = titles[i],
                                            fontsize = fontsize, legend=False)
    # Add annotation letter at the top of each column
    fig.text(i * 0.335, 0.98, annotation[i], fontsize=fontsize*1.5, fontweight='bold', ha='center', va='center')

        
fig.legend(handles, labels, loc='upper center', 
           bbox_to_anchor=(0.5, -0.1), ncol=4, fontsize=fontsize, frameon=False)

fig.tight_layout()
plt.subplots_adjust(bottom=0.005)  # Adjust this value to avoid overlap with the legend

plt.savefig(RESULT_FIG_FILE, bbox_inches='tight') #TODO: does not save legend
fig.show()
```
